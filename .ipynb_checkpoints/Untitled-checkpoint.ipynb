```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

df = pd.read_csv("train.csv") #Reading the dataset in a dataframe using Pandas
df.head()

df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Self_Employed'].fillna('No',inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mode()[0], inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)

var_mod = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
df.dtypes
df.head()

df.iloc[:, 1:12]
df.loc[:, ['Loan_Status']]

x = df.iloc[:, 1:12]
y = df.loc[:, ['Loan_Status']]

# Set the size of the plot
plt.figure(figsize=(14,7))
 
# Create a colormap
colormap = np.array(['red', 'lime'])
 
# Plot Income
plt.subplot(1, 2, 1)
plt.scatter(x.ApplicantIncome, x.CoapplicantIncome, c=colormap[y.Loan_Status], s=40)
plt.title('Income')

# Plot Loan
plt.subplot(1, 2, 2)
plt.scatter(x.ApplicantIncome, x.LoanAmount, c=colormap[y.Loan_Status], s=40)
plt.title('Loan')
```
